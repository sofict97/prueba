```python
#Devuelve los estudiantes que no tienen un dato en Grade como NaN 
df_students[df_students.isnull().any(axis=1)]
```

```python
#Muestra la lista completa incluyendo los NaN respectivamente
df_students.StudyHours = df_students.StudyHours.fillna(df_students.StudyHours.mean())
df_students
```

```python
#Muestra la lista sin los valores que no contienen valores
df_students = df_students.dropna(axis=0, how='any')
df_students
```

```python
# Obtenga el promedio de study hours usando el nombre de la columna como indice
mean_study = df_students['StudyHours'].mean()

# Obtenga el promedio grade usando el nombre de la columna como una propiedad (¬°solo para aclarar el punto!)
mean_grade = df_students.Grade.mean()

# Print el promedio de study hours y grade
print('Average weekly study hours: {:.2f}\nAverage grade: {:.2f}'.format(mean_study, mean_grade))
```

```python
# Saca el promedio de grade
df_students[df_students.StudyHours > mean_study].Grade.mean()
```

```python
#Nos muestra qui√©n aprob√≥ el curso 
passes  = pd.Series(df_students['Grade'] >= 60)
df_students = pd.concat([df_students, passes.rename("Pass")], axis=1)

df_students
```

```python
# Create a DataFrame with the data sorted by Grade (descending)
df_students = df_students.sort_values('Grade', ascending=False)

# Show the DataFrame
df_students
```

```python
#Puede agregar varios campos en un grupo utilizando cualquier funci√≥n de agregaci√≥n disponible. Por ejemplo, puede encontrar el tiempo medio de estudio y la calificaci√≥n de los grupos de estudiantes que aprobaron y reprobaron el curso.
print(df_students.groupby(df_students.Pass)['StudyHours', 'Grade'].mean())
```

```python
#puede usar el m√©todo groupby para agrupar los datos. Determinar cu√°ntos estudiantes aprobaron y reprobaron
print(df_students.groupby(df_students.Pass).Name.count())
```
